```python
import matplotlib.pyplot as plt

# Tentukan ukuran canvas
canvas_width = 10
canvas_height = 10

# Membuat canvas kosong dengan latar belakang putih
canvas = plt.figure(figsize=(canvas_width, canvas_height))
ax = canvas.add_subplot(111)

# Matikan axis (sumbu)
ax.axis('off')

# Gambar titik piksel di koordinat (5, 5)
ax.plot(5, 5, 'ro') # 'ro' berarti titik merah

# Tampilkan gambar
plt.show()
```

```python
angka_pertama = 10  # Mendefinisikan variabel pertama sebagai bilangan bulat
angka_kedua = 5  # Mendefinisikan variabel kedua sebagai bilangan bulat

hasil_penjumlahan = angka_pertama + angka_kedua  # Melakukan operasi penjumlahan antara kedua variabel

print("=== Hasil Operasi ===")  # Mencetak teks sebagai judul output
print(hasil_penjumlahan)  # Mencetak hasil dari operasi penjumlahan ke konsol
```
