```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle

# Parameter persegi panjang (x, y, width, height)
x0, y0 = 1, 1
w, h = 8, 5
points_per_side = 20

# Titik-titik di setiap sisi
top = np.column_stack((np.linspace(x0, x0 + w, points_per_side), np.full(points_per_side, y0 + h)))
bottom = np.column_stack((np.linspace(x0, x0 + w, points_per_side), np.full(points_per_side, y0)))
left = np.column_stack((np.full(points_per_side, x0), np.linspace(y0, y0 + h, points_per_side)))
right = np.column_stack((np.full(points_per_side, x0 + w), np.linspace(y0, y0 + h, points_per_side)))

fig, ax = plt.subplots(figsize=(8, 6))

# Gambar outline persegi panjang
ax.add_patch(Rectangle((x0, y0), w, h, fill=False, linewidth=1))

# Plot titik dengan warna berbeda untuk tiap sisi
ax.scatter(top[:, 0], top[:, 1], color='red',   s=30, label='Atas')
ax.scatter(bottom[:, 0], bottom[:, 1], color='blue',  s=30, label='Bawah')
ax.scatter(left[:, 0], left[:, 1], color='green', s=30, label='Kiri')
ax.scatter(right[:, 0], right[:, 1], color='orange', s=30, label='Kanan')

ax.set_aspect('equal')
ax.set_xlim(x0 - 1, x0 + w + 1)
ax.set_ylim(y0 - 1, y0 + h + 1)
ax.axis('off')
ax.legend()
plt.show()
```
