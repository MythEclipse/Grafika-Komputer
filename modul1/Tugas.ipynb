```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle

# Parameter persegi panjang (x, y, width, height)
x0, y0 = 1, 1
w, h = 8, 5

# Satu titik (tengah) pada setiap sisi
top = np.array([[x0 + w/2, y0 + h]])
bottom = np.array([[x0 + w/2, y0]])
left = np.array([[x0, y0 + h/2]])
right = np.array([[x0 + w, y0 + h/2]])

fig, ax = plt.subplots(figsize=(8, 6))

# Gambar outline persegi panjang
ax.add_patch(Rectangle((x0, y0), w, h, fill=False, linewidth=1))

# Plot titik dengan warna berbeda untuk tiap sisi
ax.scatter(top[:, 0], top[:, 1], color='red',   s=80, label='Atas')
ax.scatter(bottom[:, 0], bottom[:, 1], color='blue',  s=80, label='Bawah')
ax.scatter(left[:, 0], left[:, 1], color='green', s=80, label='Kiri')
ax.scatter(right[:, 0], right[:, 1], color='orange', s=80, label='Kanan')

ax.set_aspect('equal')
ax.set_xlim(x0 - 1, x0 + w + 1)
ax.set_ylim(y0 - 1, y0 + h + 1)
ax.axis('off')
ax.legend()
plt.show()
```
