```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots()

# Roda
roda1 = patches.Circle((1.5, 0.5), 0.5, facecolor='grey', edgecolor='black', linewidth=2)
roda2 = patches.Circle((4.5, 0.5), 0.5, facecolor='grey', edgecolor='black', linewidth=2)

# Badan mobil
badan = patches.Rectangle((0, 1), 6, 2, facecolor='skyblue', edgecolor='blue', linewidth=2)

# Kabin mobil
kabin = patches.Rectangle((1, 2.5), 4, 1, facecolor='skyblue', edgecolor='blue', linewidth=2)

ax.add_patch(badan)
ax.add_patch(kabin)
ax.add_patch(roda1)
ax.add_patch(roda2)

# Mengatur batas-batas tampilan sumbu
ax.set_xlim(-1, 7)
ax.set_ylim(-1, 4)

# Mengatur aspek rasio agar lingkaran tidak lonjong
ax.set_aspect('equal')

plt.grid(True)
plt.show()
```
