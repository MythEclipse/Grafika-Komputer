```python
import matplotlib.pyplot as plt
import math

# Panjang sisi
s = 10
# Tinggi segitiga sama sisi
h = s * math.sqrt(3) / 2

# Koordinat segitiga (kembali ke titik awal untuk menutup segitiga)
x = [0, s, s/2, 0]
y = [0, 0, h, 0]

plt.figure()
plt.plot(x, y, color='red', linewidth=2)  # garis merah
plt.gca().set_aspect('equal', 'box')
plt.title('Segitiga sama sisi - sisi = 10')
plt.xlim(-1, s+1)
plt.ylim(-1, h+1)
plt.grid(False)
plt.show()
```

```python
# Gambar dua persegi panjang yang saling menyatu
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots()
# Persegi panjang pertama (x=1, y=1), lebar=4, tinggi=2
rect1 = patches.Rectangle((1, 1), 4, 2, linewidth=2, edgecolor='blue', facecolor='none')
# Persegi panjang kedua menyatu pada sisi kanan persegi panjang pertama (x=5, y=1), lebar=3, tinggi=2
rect2 = patches.Rectangle((5, 1), 3, 2, linewidth=2, edgecolor='green', facecolor='none')

ax.add_patch(rect1)
ax.add_patch(rect2)
ax.set_aspect('equal', 'box')
ax.set_xlim(0, 9)
ax.set_ylim(0, 4)
ax.set_title('Dua persegi panjang yang saling menyatu')
ax.grid(False)
plt.show()
```
