```python
import matplotlib.pyplot as plt

# --- Fungsi DDA (Digital Differential Analyzer) ---
def dda_algorithm(x0, y0, x1, y1):
    x, y = x0, y0
    dx = x1 - x0
    dy = y1 - y0
    
    # Menentukan 'steps' sebagai nilai absolut terbesar antara dx dan dy
    steps = abs(dx) if abs(dx) > abs(dy) else abs(dy)
    
    # Menghitung inkremen (perubahan per langkah)
    X_inc = dx / steps
    Y_inc = dy / steps
    
    # Inisialisasi list untuk menyimpan titik koordinat
    x_points = [x]
    y_points = [y]
    
    # Loop untuk menghasilkan titik-titik garis
    for i in range(int(steps)):
        x += X_inc
        y += Y_inc
        
        # Simpan koordinat setelah dibulatkan
        x_points.append(round(x))
        y_points.append(round(y))
        
    # Plot garis DDA
    plt.plot(x_points, y_points, marker="o", color="b")

# --- Fungsi Menggambar Rumah ---
def draw_house():
    # Mengatur ukuran plot
    plt.figure(figsize=(6, 6))

    # Dinding rumah (persegi) - Tidak ada perubahan
    dda_algorithm(2, 2, 2, 6) # Kiri
    dda_algorithm(2, 6, 6, 6) # Atas
    dda_algorithm(6, 6, 6, 2) # Kanan
    dda_algorithm(6, 2, 2, 2) # Bawah

    # Atap rumah (segitiga) - Tidak ada perubahan
    dda_algorithm(2, 6, 4, 8) # Kiri ke puncak
    dda_algorithm(4, 8, 6, 6) # Puncak ke Kanan

    # Pintu (persegi kecil) - Bergeser dan lebih lebar
    # Koordinat pintu di gambar baru: (3,2) ke (3,4) dan (3,4) ke (4,4) dan (4,4) ke (4,2)
    dda_algorithm(3, 2, 3, 4) # Kiri
    dda_algorithm(3, 4, 4, 4) # Atas
    dda_algorithm(4, 4, 4, 2) # Kanan
    
    # Jendela Kiri (persegi) - Baru
    # Koordinat: (2,4) ke (2,5), (2,5) ke (3,5), (3,5) ke (3,4), (3,4) ke (2,4)
    dda_algorithm(2, 4, 2, 5) # Kiri
    dda_algorithm(2, 5, 3, 5) # Atas
    dda_algorithm(3, 5, 3, 4) # Kanan
    dda_algorithm(3, 4, 2, 4) # Bawah

    # Jendela Kanan (persegi) - Baru
    # Koordinat: (5,4) ke (5,5), (5,5) ke (6,5), (6,5) ke (6,4), (6,4) ke (5,4)
    dda_algorithm(5, 4, 5, 5) # Kiri
    dda_algorithm(5, 5, 6, 5) # Atas
    dda_algorithm(6, 5, 6, 4) # Kanan
    dda_algorithm(6, 4, 5, 4) # Bawah

    # Menampilkan hasil gambar rumah
    plt.title("Gambar Rumah dengan Algoritma DDA")
    plt.xlim(0, 8)          # Batas sumbu X dari 0 sampai 8
    plt.ylim(0, 10)         # Batas sumbu Y dari 0 sampai 10
    
    # Mengatur rasio aspek (agar gambar tidak terdistorsi)
    plt.gca().set_aspect('equal', adjustable='box') 
    plt.grid(True)
    plt.show()

# --- Memanggil fungsi untuk menggambar ---
draw_house()
```
