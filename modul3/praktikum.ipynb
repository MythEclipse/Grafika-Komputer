```python
import matplotlib.pyplot as plt
import numpy as np

def garis_dda(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0
    step = max(abs(dx), abs(dy))
    Xinc = dx / step
    Yinc = dy / step
    x = x0
    y = y0
    x_points = []
    y_points = []

    for i in range(int(step) + 1):
        x_points.append(int(round(x)))
        y_points.append(int(round(y)))
        x += Xinc
        y += Yinc

    plt.plot(x_points, y_points, marker='o', color='blue')
    plt.title('Garis dengan Algoritma DDA')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.show()

garis_dda(2, 3, 10, 8)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def dda_algorithm(x0, y0, x1, y1):
    # Menghitung perbedaan koordinat
    dx = x1 - x0
    dy = y1 - y0

    # Menentukan jumlah langkah berdasarkan jarak terpanjang (step)
    steps = max(abs(dx), abs(dy))

    # Menghitung perubahan x dan y per langkah
    # Menggunakan float(steps) untuk memastikan pembagian menghasilkan nilai float
    Xinc = dx / float(steps)
    Yinc = dy / float(steps)

    # Inisialisasi titik awal
    x = x0
    y = y0

    # Menyimpan titik-titik hasil algoritma DDA
    x_points = [x]
    y_points = [y]

    # Loop untuk menghitung titik-titik sepanjang garis
    # Loop sebanyak 'steps' kali (tidak termasuk titik awal yang sudah disimpan)
    for i in range(int(steps)):
        x += Xinc
        y += Yinc
        
        # Pembulatan ke bilangan bulat terdekat sebelum disimpan
        x_points.append(round(x))
        y_points.append(round(y))

    # Menggambar garis menggunakan matplotlib
    plt.plot(x_points, y_points, marker='o', color='b')
    plt.title(f"Garis DDA dari ({x0}, {y0}) ke ({x1}, {y1})")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()

# --- INPUT PENGGUNA ---

# Meminta input dari pengguna untuk titik awal dan titik akhir
try:
    x0 = int(input("Masukkan koordinat x0 (titik awal): "))
    y0 = int(input("Masukkan koordinat y0 (titik awal): "))
    x1 = int(input("Masukkan koordinat x1 (titik akhir): "))
    y1 = int(input("Masukkan koordinat y1 (titik akhir): "))

    # Memanggil fungsi DDA untuk menggambar garis
    dda_algorithm(x0, y0, x1, y1)

except ValueError:
    print("Input tidak valid. Pastikan semua koordinat adalah bilangan bulat.")
```

```python
import matplotlib.pyplot as plt

def bresenham_algorithm(x0, y0, x1, y1):
    # Inisialisasi variabel
    x, y = x0, y0
    dx = abs(x1 - x0)
    dy = abs(y1 - y0)
    
    # Menentukan arah penambahan (increment) untuk x dan y
    sx = 1 if x0 < x1 else -1  # penambahan untuk x
    sy = 1 if y0 < y1 else -1  # penambahan untuk y
    
    # Inisialisasi parameter keputusan (error)
    # Ini adalah versi Bresenham yang dimodifikasi untuk menangani semua oktan (sudut)
    err = dx - dy

    # Menyimpan titik awal
    x_points = [x]
    y_points = [y]

    # Loop utama algoritma
    # Berlanjut selama titik saat ini (x, y) belum mencapai titik akhir (x1, y1)
    while x != x1 or y != y1:
        # Menghitung parameter keputusan berikutnya
        e2 = 2 * err
        
        # Periksa apakah piksel berikutnya harus bergerak di sumbu Y
        if e2 > -dy:
            err -= dy
            x += sx
            
        # Periksa apakah piksel berikutnya harus bergerak di sumbu X
        if e2 < dx:
            err += dx
            y += sy
            
        # Menyimpan koordinat piksel baru
        x_points.append(x)
        y_points.append(y)

    # Menggambar garis menggunakan matplotlib
    plt.plot(x_points, y_points, marker="o", color="r")
    plt.title(f"Garis Bresenham dari ({x0}, {y0}) ke ({x1}, {y1})")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()

# --- INPUT PENGGUNA ---

# Meminta input dari pengguna untuk titik awal dan titik akhir
try:
    x0 = int(input("Masukkan koordinat x0 (titik awal): "))
    y0 = int(input("Masukkan koordinat y0 (titik awal): "))
    x1 = int(input("Masukkan koordinat x1 (titik akhir): "))
    y1 = int(input("Masukkan koordinat y1 (titik akhir): "))

    # Memanggil fungsi untuk menggambar garis dengan algoritma Bresenham
    bresenham_algorithm(x0, y0, x1, y1)

except ValueError:
    print("Input tidak valid. Pastikan semua koordinat adalah bilangan bulat.")
```

```python
import matplotlib.pyplot as plt

# --- Fungsi DDA (Digital Differential Analyzer) ---
def dda_algorithm(x0, y0, x1, y1):
    x, y = x0, y0
    dx = x1 - x0
    dy = y1 - y0
    
    # Menentukan 'steps' sebagai nilai absolut terbesar antara dx dan dy
    steps = abs(dx) if abs(dx) > abs(dy) else abs(dy)
    
    # Menghitung inkremen (perubahan per langkah)
    X_inc = dx / steps
    Y_inc = dy / steps
    
    # Inisialisasi list untuk menyimpan titik koordinat
    x_points = [x]
    y_points = [y]
    
    # Loop untuk menghasilkan titik-titik garis
    for i in range(int(steps)):
        x += X_inc
        y += Y_inc
        
        # Simpan koordinat setelah dibulatkan
        x_points.append(round(x))
        y_points.append(round(y))
        
    # Plot garis DDA
    # Garis akan diplot di dalam fungsi DDA, dan fungsi draw_house akan memanggilnya
    plt.plot(x_points, y_points, marker="o", color="b")

# --- Fungsi Menggambar Rumah ---
def draw_house():
    # Mengatur ukuran plot
    plt.figure(figsize=(6, 6))

    # Dinding rumah (persegi)
    # Kiri: (2, 2) ke (2, 6)
    dda_algorithm(2, 2, 2, 6) 
    # Atas: (2, 6) ke (6, 6)
    dda_algorithm(2, 6, 6, 6) 
    # Kanan: (6, 6) ke (6, 2)
    dda_algorithm(6, 6, 6, 2) 
    # Bawah: (6, 2) ke (2, 2)
    dda_algorithm(6, 2, 2, 2) 

    # Atap rumah (segitiga)
    # Kiri ke puncak: (2, 6) ke (4, 8)
    dda_algorithm(2, 6, 4, 8) 
    # Puncak ke Kanan: (4, 8) ke (6, 6)
    dda_algorithm(4, 8, 6, 6) 

    # Pintu (persegi kecil)
    # Kiri: (3, 2) ke (3, 4)
    dda_algorithm(3, 2, 3, 4) 
    # Atas: (3, 4) ke (4, 4)
    dda_algorithm(3, 4, 4, 4) 
    # Kanan: (4, 4) ke (4, 2)
    dda_algorithm(4, 4, 4, 2) 
    
    # Jendela (persegi kecil)
    # Kiri: (5, 4) ke (5, 5)
    dda_algorithm(5, 4, 5, 5) 
    # Atas: (5, 5) ke (6, 5) 
    dda_algorithm(5, 5, 6, 5) 
    # Kanan: (6, 5) ke (6, 4)
    dda_algorithm(6, 5, 6, 4) 
    # Bawah: (6, 4) ke (5, 4)
    dda_algorithm(6, 4, 5, 4) 

    # Menampilkan hasil gambar rumah
    plt.title("Gambar Rumah dengan Algoritma DDA")
    plt.xlim(0, 8)          # Batas sumbu X dari 0 sampai 8
    plt.ylim(0, 10)         # Batas sumbu Y dari 0 sampai 10
    
    # Mengatur rasio aspek (agar gambar tidak terdistorsi)
    plt.gca().set_aspect('equal', adjustable='box') 
    plt.grid(True)
    plt.show()

# --- Memanggil fungsi untuk menggambar ---
draw_house()
```
