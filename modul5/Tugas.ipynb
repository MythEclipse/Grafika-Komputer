```python
import matplotlib.pyplot as plt
import numpy as np

x_awal = [10, 40, 50, 20, 10]
y_awal = [10, 10, 20, 20, 10]

x_translasi = [20, 50, 60, 30, 20]
y_translasi = [20, 20, 30, 30, 20]

plt.figure(figsize=(8, 5))

plt.plot(x_awal, y_awal, color='black', linewidth=2.5, label='Persegi Panjang Awal')
plt.plot(x_translasi, y_translasi, color='black', linewidth=1, label='Persegi Panjang Setelah Translasi')

plt.xlim(0, 65)
plt.ylim(0, 35)

plt.xticks(np.arange(0, 70, 10))
plt.yticks(np.arange(0, 40, 5))

plt.grid(True)
plt.legend(loc='lower right', frameon=True, facecolor='white', framealpha=1)
plt.title('Translasi Jajargenjang Bertingkat')

plt.show()
```
